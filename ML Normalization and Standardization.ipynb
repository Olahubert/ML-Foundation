```python
#iport sklearn
from sklearn.preprocessing import MinMaxScaler
```

```python
#create the scaler variable
scaler = MinMaxScaler()
```

```python
# to Normalization
```

```python
df1 = pd.DataFrame(scaler.fit_transform(TIP[["total_bill"]]))
df1
```

```python
#Add the name of the column and index
df1 = pd.DataFrame(scaler.fit_transform(TIP[["total_bill"]]), columns=["normalized_total_bill"], index=TIP.index)
df1
```

```python
#Normalizing multiple column
```

```python
df2 = pd.DataFrame(scaler.fit_transform(Tip_Num), columns=["normalized_total_bill","normalized_tip", "normalized_size"], index=TIP.index)
df2
```

```python
# Add the column to the dataframe
```

```python
TIP['normalized_total_bill'] = df2.normalized_total_bill
```

```python
TIP
```

```python
TIP['normalized_tip'] = df2.normalized_tip
TIP
```

```python
TIP['normalized_size'] = df2.normalized_size
TIP
```
