```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
Titanic['new_sex'] = le.fit_transform(Titanic['sex'])
```

```python
Titanic['new_class'] = le.fit_transform(Titanic['class'])
```

```python
Titanic['new_embarked'] = le.fit_transform(Titanic['embarked'])
```

```python
Titanic['new_embark_town'] = le.fit_transform(Titanic['embark_town'])
```

```python
Titanic['new_alive'] = le.fit_transform(Titanic['alive'])
```
