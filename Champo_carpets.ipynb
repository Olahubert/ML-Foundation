```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel('Champ_Carpets2.xlsx')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df
```

```python
numerical_cols = ['QtyRequired', 'TotalArea', 'Amount']
for col in numerical_cols:
    print(f"Descriptive Statistics for {col}:")
    print(df[col].describe())
```

```python
categorical_cols = ['OrderType', 'OrderCategory', 'CustomerCode', 'CountryName', 'CustomerOrderNo', 'Custorderdate', 'UnitName', 'QualityName', 'DesignName', 'ColorName', 'ShapeName']
for col in categorical_cols:
    print(f"Unique Values for {col}:")
    print(df[col].unique())
```

```python
plt.figure(figsize=(17,6))
sns.countplot(x='CountryName', data=df)
plt.title('Distribution of Orders by Country')
plt.savefig('Champ_Carpets5.png')
plt.show()
```

```python
sns.scatterplot(data=df, x='TotalArea', y='Amount')
plt.title('Order Amount vs Total Area')
plt.savefig('Champ_Carpets3.png')
plt.show()
```

```python
df['Custorderdate'] = pd.to_datetime(df['Custorderdate'])
df['Year'] = df['Custorderdate'].dt.year
sns.lineplot(data=df, x='Year', y='QtyRequired')
plt.title('Quantity Required Over Time')
plt.savefig('Champ_Carpets4.png')
plt.show()
```
